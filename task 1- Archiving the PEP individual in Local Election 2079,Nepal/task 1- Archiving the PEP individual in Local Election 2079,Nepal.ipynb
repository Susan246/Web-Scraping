### Data scraping using BeautifulSoup, Selenium and Python

```python
# importing necessary libraries
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException,StaleElementReferenceException, ElementClickInterceptedException
from selenium.webdriver.chrome.options import Options
import pandas as pd
from bs4 import BeautifulSoup


# Initialize WebDriver 
driver = webdriver.Chrome(service=Service("D:/pythonprogramme/selenium/chromedriver.exe"))

# Loadding page
driver.get('https://localelection2079.ekantipur.com/?lng=eng') 


# function for ward block
def select_wards_with_retry(retries=10):
    attempt = 0
    while attempt < retries:
        if driver.find_element(By.CSS_SELECTOR, '#c > a').text == 'Local Election 2079':
            try:
                # Ward Selection
                WebDriverWait(driver, 10).until(
                    EC.presence_of_all_elements_located((By.XPATH, '//*[@id="select-box"]/option'))
                )

                # ward data scraping
                page_source = driver.page_source
                soup = BeautifulSoup(page_source, "html.parser")

                _country = "Nepal"
                _election_name = "Local Election 2079"
                _province = soup.find_all('li', class_='breadcrumb-item')[1].text.strip()
                _district = soup.find_all('li', class_='breadcrumb-item')[2].text.strip()
                _municipality = soup.find_all('li', class_='breadcrumb-item')[3].text.strip()

                print('Successful')
                
                for tab_container in soup.find('div', class_="tab-container"):
                    j = 0
                    
                    for wada in tab_container.find_all('div', class_='tab-content'):
                        total_wards = len(driver.find_elements(By.XPATH, '//*[@id="select-box"]/option'))
                        _ward = j + 1
                        
                        position_boxes = wada.find_all('div', class_='col-md-6 col-lg-3')

                        # ward president block
                        for box1 in position_boxes[0]:
                            try:
                                _position = box1.find('div', class_='card-header').text.strip()
                                _name = box1.find('div', class_='candidate-name').text.strip()
                                _party = box1.find('div', class_='candidate-party-name').text.strip()
                                _votes = box1.find('div', class_='vote-numbers').text.strip()
                                candidates.append([_country, _election_name, _province, _district, _municipality, _name, _position, _ward, _party, _votes])
                            except AttributeError:
                                break

                        # female member block
                        for box2 in position_boxes[1]:
                            try:
                                _position = box2.find('div', class_='card-header').text.strip()
                                _name = box2.find('div', class_='candidate-name').text.strip()
                                _party = box2.find('div', class_='candidate-party-name').text.strip()
                                _votes = box2.find('div', class_='vote-numbers').text.strip()
                                candidates.append([_country, _election_name, _province, _district, _municipality, _name, _position, _ward, _party, _votes])
                            except AttributeError:
                                break

                        # dalit female member block
                        for box3 in position_boxes[2]:
                            try:
                                _position = box3.find('div', class_='card-header').text.strip()
                                _name = box3.find('div', class_='candidate-name').text.strip()
                                _party = box3.find('div', class_='candidate-party-name').text.strip()
                                _votes = box3.find('div', class_='vote-numbers').text.strip()
                                candidates.append([_country, _election_name, _province, _district, _municipality, _name, _position, _ward, _party, _votes])
                            except AttributeError:
                                break

                        # member block
                        for box4 in position_boxes[3]:
                            _position = box4.find('div', class_='card-header').text.strip()
                            for cube in box4.find_all('div', class_='candidate-meta-wrapper')[:2]:
                                try:
                                    _name = cube.find('div', class_='candidate-name').text.strip()
                                    _party = cube.find('div', class_='candidate-party-name').text.strip()
                                    _votes = cube.find('div', class_='vote-numbers').text.strip()
                                    candidates.append([_country, _election_name, _province, _district, _municipality, _name, _position, _ward, _party, _votes])
                                except AttributeError:
                                    break

                        if j < total_wards:
                            j += 1

                break

            except TimeoutException:
                attempt += 1
                while True:
                    try:
                        search_button = driver.find_element(By.XPATH, '//*[@id="search-result"]')
                        driver.execute_script("arguments[0].scrollIntoView(true);", search_button)
                        WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="search-result"]')))
                        driver.execute_script("arguments[0].click();", search_button)

                        WebDriverWait(driver, 20).until(
                            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "body > div.container > div.card.p-3.my-4 > div > div.col-xs-12.col-lg > div.card-body.section-1"))
                        )

                        page_source = driver.page_source
                        soup = BeautifulSoup(page_source, "html.parser")
                        for candidate in soup.find_all('div', class_='card border-0'):
                            position_ = candidate.find('h6', class_='card-title fw-bold').text
                            for can in candidate.find_all('li', class_='list-group-item px-0 election-list winning-candidate'):
                                country_ = "Nepal"
                                election_name_ = "Local Election 2079"
                                province_ = soup.find_all('li', class_='breadcrumb-item')[1].text.strip()
                                district_ = soup.find_all('li', class_='breadcrumb-item')[2].text.strip()
                                municipality_ = soup.find_all('li', class_='breadcrumb-item')[3].text.strip()
                                name_ = can.find('div', class_='candidate-name').text
                                ward_no_ = ""
                                party_ = can.find('div', class_='candidate-party-name').text
                                votes_ = can.find('div', class_='vote-numbers').text

                                candidates.append([country_, election_name_, province_, district_, municipality_, name_, position_, ward_no_, party_, votes_])


                        break
                    except (ElementClickInterceptedException, StaleElementReferenceException, TimeoutException):
                        continue

                if attempt >= retries:
                    print("Failed to load ward options after multiple retries.")
                    break

        else:
            driver.execute_script("window.history.go(-1)")
                    
# function for skipped pages
def retry_skipped_pages(skipped_pages, max_retries=5):
    for page in skipped_pages:
        attempt = 0
        success = False
        while attempt < max_retries and not success:
            try:
                driver.get(page)
                WebDriverWait(driver, 10).until(
                    EC.presence_of__element_located((By.CSS_SELECTOR, "body > div.container > div.card.p-3.my-4 > div > div.col-xs-12.col-lg > div.card-body.section-1"))
                )
                
                page_source = driver.page_source
                soup = BeautifulSoup(page_source, "html.parser")

                for candidate in soup.find_all('div', class_='card border-0'):
                    position_ = candidate.find('h6', class_='card-title fw-bold').text
                    for can in candidate.find_all('li', class_='list-group-item px-0 election-list winning-candidate'):
                        country_ = "Nepal"
                        election_name_ = "Local Election 2079"
                        province_ = soup.find_all('li', class_='breadcrumb-item')[1].text.strip()
                        district_ = soup.find_all('li', class_='breadcrumb-item')[2].text.strip()
                        municipality_ = soup.find_all('li', class_='breadcrumb-item')[3].text.strip()
                        name_ = can.find('div', class_='candidate-name').text
                        ward_no_ = ""
                        party_ = can.find('div', class_='candidate-party-name').text
                        votes_ = can.find('div', class_='vote-numbers').text
                        
                        candidates.append([country_, election_name_, province_, district_, municipality_, name_, position_, ward_no_, party_, votes_])
                        print("This is from skipped page at last")
                
                select_wards_with_retry()
                
                success = True

            except (TimeoutException, StaleElementReferenceException, NoSuchElementException):
                attempt += 1
                print(f"Retrying {page} ({attempt}/{max_retries})")

            if attempt >= max_retries:
                print(f"Failed to retrieve {page} after {max_retries} attempts.")

candidates=[]                
skipped_pages=[]
pro_index = 0
dist_index = 0
muni_index = 0
count=0

province_options = driver.find_elements(By.XPATH, "//*[@id='provinceDD']/option")[1:]
while pro_index < len(province_options):
    try:
        WebDriverWait(driver, 120).until(
            EC.presence_of_all_elements_located((By.XPATH, '//*[@id="provinceDD"]/option'))
        )
    
        # Province Selection
        province_options = driver.find_elements(By.XPATH, "//*[@id='provinceDD']/option")[1:]
        
        print("province:",province_options[pro_index].text)
        
        province_options[pro_index].click()
        time.sleep(1)
        
        # District Selection
        district_options = driver.find_elements(By.XPATH, '//*[@id="districtDD"]/option')[1:]
        while dist_index < len(district_options):
            district_options = driver.find_elements(By.XPATH, '//*[@id="districtDD"]/option')[1:]
            
            print("distirct:",district_options[dist_index].text)
            
            district_options[dist_index].click()
            time.sleep(1)
            
            # Municipality Selection
            municipality_options = driver.find_elements(By.XPATH, '//*[@id="locationDD"]/option')[1:]
            while muni_index < len(municipality_options):
                municipality_options = driver.find_elements(By.XPATH, '//*[@id="locationDD"]/option')[1:]
                
                print("municipality:",municipality_options[muni_index].text)
                
                municipality_options[muni_index].click()
                time.sleep(1)
                
                while True:
                    try:
                        search_button = driver.find_element(By.XPATH, '//*[@id="search-result"]')
                        driver.execute_script("arguments[0].scrollIntoView(true);", search_button)
                        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="search-result"]')))
                        driver.execute_script("arguments[0].click();", search_button)
                        break
                    except (ElementClickInterceptedException, StaleElementReferenceException):
                        continue
                
                page_source = driver.page_source
                soup = BeautifulSoup(page_source, "html.parser")
                
                for candidate in soup.find_all('div', class_='card border-0'):
                    position_ = candidate.find('h6', class_='card-title fw-bold').text
                    for can in candidate.find_all('li', class_='list-group-item px-0 election-list winning-candidate'):
                        country_ = "Nepal"
                        election_name_ = "Local Election 2079"
                        province_ = soup.find_all('li', class_='breadcrumb-item')[1].text.strip()
                        district_ = soup.find_all('li', class_='breadcrumb-item')[2].text.strip()
                        municipality_ = soup.find_all('li', class_='breadcrumb-item')[3].text.strip()
                        name_ = can.find('div', class_='candidate-name').text
                        ward_no_ = ""
                        party_ = can.find('div', class_='candidate-party-name').text
                        votes_ = can.find('div', class_='vote-numbers').text
                        
                        candidates.append([country_, election_name_, province_, district_, municipality_, name_, position_, ward_no_, party_, votes_])
                        
                
                # calling ward function
                select_wards_with_retry()
                
                muni_index += 1

            dist_index += 1
            muni_index = 0

        pro_index += 1
        dist_index = 0
        muni_index = 0

    except (NoSuchElementException, AttributeError):
        skipped_pages.append(driver.current_url)
        print("This page have some issues and will be retry at last")
        print(driver.current_url)
        
        if muni_index < len(municipality_options):
            muni_index += 1
        elif dist_index < len(district_options):
            dist_index += 1
            muni_index = 0
        else:
            pro_index += 1
            dist_index = 0
            muni_index = 0
        
        driver.execute_script("window.history.go(-1)")

# Calling the retry function for skipped pages
retry_skipped_pages(skipped_pages)

driver.quit()
```

### Creating dataframe 

```python
df=pd.DataFrame(candidates,columns=['country','election name','province','district','municipality','name','position','ward no','party name','received votes'])
```

## Transfering data into database

```python
insert_query = """
INSERT INTO vote_details (country, `election name`, province, district, municipality, `name`, `position`, `ward no`, `party name`, `received votes`)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for i, row in df.iterrows():
    cursor.execute(insert_query, (
        row['country'],
        row['election name'],
        row['province'],
        row['district'],
        row['municipality'],
        row['name'],
        row['position'],
        row['ward no'],
        row['party name'],
        row['received votes']
    ))

# Commit the transaction
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```
